```python
# 1. Get user input and convert it to an integer (number)
score_text = input("Enter AI confidence score (0-100): ")
score = int(score_text)

# 2. Check the score against thresholds
if score >= 80:
    print("High confidence - Accept.")
    
elif score >= 50:
    # We don't need to check "score < 80" here because 
    # the first 'if' already handled everything 80 and above.
    print("Medium - Review.")
    
else:
    # This catches everything below 50
    print("Low - Reject.")
```
