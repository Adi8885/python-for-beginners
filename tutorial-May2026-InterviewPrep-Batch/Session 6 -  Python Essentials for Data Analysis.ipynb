### 1. Selecting Columns 🥞

### 2. Filtering Rows (Boolean Indexing) 🕵️‍♂️
You can filter rows based on conditions. The condition inside the brackets returns `True` or `False` for each row, and Pandas keeps only the `True` rows.

```python
# 1. Single condition (Age greater than 25)
df[df['age'] > 25]

# 2. Exact match (City is Mumbai)
df[df['city'] == 'Mumbai']

# 3. AND Condition (&): Both conditions must be True
df[(df['age'] > 25) & (df['city'] == 'Mumbai')]

# 4. OR Condition (|): At least one condition must be True
df[(df['age'] < 25) | (df['age'] > 60)]
```
> ⚠️ **Important Syntax Rule:** When combining filters in Pandas, **always** wrap each condition in parentheses `()` and use `&` (AND) or `|` (OR) instead of the plain python `and` / `or` words.

---

### 3. Label-based (`loc`) vs Position-based (`iloc`) Selection 📍

This is one of the most common interview topics! Let's clear the confusion once and for all:

| Feature | `loc` (Label-Based) 🏷️ | `iloc` (Integer/Index-Based) 🔢 |
| :--- | :--- | :--- |
| **How it finds data** | Uses **Row labels** and **Column names**. | Uses **Integer positions** (0, 1, 2...). |
| **End Index** | **Inclusive** (includes the stop value). | **Exclusive** (excludes the stop value, like standard python slices). |
| **Example Code** | `df.loc[0:5, ['name', 'age']]` | `df.iloc[0:5, 0:3]` |
| **What it grabs** | Rows labeled 0 to 5 (6 rows total), columns named 'name' and 'age'. | Rows at indices 0, 1, 2, 3, 4 (5 rows total), columns at index 0, 1, 2. |

---

### 4. The `query()` Method: Write cleaner code! 📝
If you find the bracket syntax `df[(df['age'] > 25) & ...]` too cluttered, you can use the SQL-like `query()` method:

```python
# Clean and readable string-based query
df.query("age > 25 and city == 'Mumbai'")
```

## 🧹 Part 1: Handling Missing Data (8 mins)

In real life, data is messy and full of empty holes (`NaN`). Pandas gives you powerful tools to detect, drop, or fill these gaps.

### 🕵️‍♂️ 1. Spotting Nulls
```python
# Check how many missing values exist in each column
df.isnull().sum()
```

### 🗑️ 2. Dropping Nulls
Sometimes it is safer to just remove rows containing missing data:

```python
df.dropna()                    # ❌ Drop any row that has at least one null value
df.dropna(subset=['column'])   # 🎯 Drop rows only if a specific column is null
df.dropna(thresh=2)            # ⚖️ Keep rows that have at least 2 non-null (valid) values
```

### 🩹 3. Filling Nulls (Imputation)
Instead of deleting data, you can replace missing values with sensible defaults:

```python
df.fillna(0)                           # 🔢 Replace all NaN values with 0
df['age'].fillna(df['age'].mean())     # 🧮 Fill age with the average age of the dataset
df['city'].fillna('Unknown')           # 🏷️ Fill text columns with a placeholder like 'Unknown'

# Propagating values
df.fillna(method='ffill')              # ⏩ Forward Fill: copies the last valid value forward
df.fillna(method='bfill')              # ⏪ Backward Fill: copies the next valid value backward
```

### 🔄 4. Replacing Custom Placeholders
If your dataset uses strings like `"N/A"`, `"missing"`, or `"?"` instead of standard `NaN` values:

```python
# Convert custom string placeholders to actual numpy Null objects
df['status'].replace('N/A', np.nan)
```

## 👥 Part 2: Groupby Operations (12 mins)

The **Split-Apply-Combine** pattern is a core concept in data analysis. Use `groupby()` to group data by categories and compute statistics.

## 💡 Part 4: Creating New Columns (10 mins)

### 2. Conditional Transformations 🚦

| Method | Best For | Speed / Performance |
| :--- | :--- | :--- |
| **`df.apply(lambda)`** | Simple single-condition custom logic. | 🐢 Slowest (processes row-by-row in Python loop). |
| **`np.where()`** | Binary choices (If / Else). | ⚡ Extremely Fast (Vectorized C-speed). |
| **`np.select()`** | Multiple conditions (If / Elif / Else). | ⚡ Extremely Fast (Vectorized C-speed). |

#### ⚙️ Option A: The standard `apply()` with lambda (or a custom function)
```python
# Simple lambda logic
df['category'] = df['age'].apply(lambda x: 'Senior' if x >= 60 else 'Adult' if x >= 18 else 'Minor')

# Defining a clean helper function first (Recommended for complex logic)
def categorize_age(age):
    if age < 18:
        return 'Minor'
    elif age < 60:
        return 'Adult'
    else:
        return 'Senior'

df['age_group'] = df['age'].apply(categorize_age)
```

#### ⚡ Option B: Binary Condition with `np.where` (Fastest!)
```python
# Syntax: np.where(condition, value_if_true, value_if_false)
df['high_value'] = np.where(df['revenue'] > 10000, 'Yes', 'No')
```

#### 🔀 Option C: Multi-Condition with `np.select` (Cleanest & Vectorized)
```python
conditions = [
    df['revenue'] < 1000,                 # Condition 1
    df['revenue'] < 5000,                 # Condition 2
    df['revenue'] >= 5000                 # Condition 3
]
choices = ['Low', 'Medium', 'High']      # Matching values

df['revenue_tier'] = np.select(conditions, choices, default='Unknown')
```

### 🎯 Key Tasks to Solve:
1. **Calculate Churn Rate**: Overall rate & rate sliced by contract type.
2. **Analyze Tenure**: Find average tenure for churned vs. retained customers.

### 🧹 Step 2: Data Cleaning & Preprocessing (6 mins)

# 🔗 Part 4: Merging & Joining DataFrames (20 mins)

In real-world databases, data is split across multiple tables to avoid duplication. Pandas uses `pd.merge()` to combine these tables, similar to SQL Joins.

---

## 🤝 1. Merge Basics (10 mins)

Here is a visual cheatsheet on how different joins combine two tables (`df1` and `df2`):

| Join Type (`how=`) | Visual Concept | What it Keeps |
| :--- | :--- | :--- |
| **`inner`** (Default) | 🔀 Intersection | Only rows with matching keys in **both** DataFrames. |
| **`left`** | 👈 Left Heavy | All rows from the left table, plus matched rows from the right table. |
| **`right`** | 👉 Right Heavy | All rows from the right table, plus matched rows from the left table. |
| **`outer`** | 🌐 Union | All rows from both tables, filling mismatches with `NaN`. |

### 🛠️ Key Merge Syntax Examples:

```python
# 1. Standard Joins on a shared column
merged = pd.merge(df1, df2, on='customer_id', how='inner')
merged = pd.merge(df1, df2, on='customer_id', how='left')
merged = pd.merge(df1, df2, on='customer_id', how='right')
merged = pd.merge(df1, df2, on='customer_id', how='outer')

# 2. When matching column names differ between tables
merged = pd.merge(
    df1, df2,
    left_on='cust_id',      # Column name in df1
    right_on='customer_id',  # Column name in df2
    how='left'
)

# 3. Joining on multiple keys (e.g., matching customer and date)
merged = pd.merge(
    df1, df2,
    on=['customer_id', 'date'],
    how='inner'
)

# 4. Custom Suffixes (resolving column name collisions)
# If both tables have a "price" column, they become "price_sales" and "price_returns"
merged = pd.merge(
    df1, df2,
    on='customer_id',
    how='left',
    suffixes=('_sales', '_returns')
)
```

# ⚡ Part 5: Quick Pandas Tricks (5 mins)

Write faster, cleaner, and more Pythonic code with these built-in tricks.

---

## 🏎️ 1. Vectorization vs. Loops (Performance Tip)
* **Rule**: NEVER use `for` loops to iterate over rows in a DataFrame.
* **Why?**: Loops process items one by one. Vectorized operations run calculations in parallel at the C level.

```python
# 🟢 Vectorized (FAST - DO THIS!)
df['new_col'] = df['col1'] * df['col2']

# 🔴 Loops (EXTREMELY SLOW - AVOID THIS!)
for i in df.index:
    df.loc[i, 'new_col'] = df.loc[i, 'col1'] * df.loc[i, 'col2']
```

---

## 🧑‍🤝‍🧑 2. Row Operations with `.apply()`
Use `axis=1` to run custom operations across columns inside a single row.

```python
df['full_name'] = df.apply(lambda row: f"{row['first_name']} {row['last_name']}", axis=1)
```

## 🔠 3. String Manipulation (`.str`)
Pandas has built-in string methods to clean raw text columns. Always prefix them with `.str`.

```python
# Extract email domain (e.g., user@gmail.com -> gmail.com)
df['email_domain'] = df['email'].str.split('@').str[1]

# Convert strings to uppercase
df['name_upper'] = df['name'].str.upper()

# Boolean search (Returns True if string contains 'gmail')
df['contains_gmail'] = df['email'].str.contains('gmail')
```

## 📅 4. DateTime Magic (`.dt`)
Parsing dates makes it easy to extract temporal features. Always convert your string date columns to datetime first!

```python
# 1. Convert column to Datetime format
df['order_date'] = pd.to_datetime(df['order_date'])

# 2. Extract components using the .dt accessor
df['year']        = df['order_date'].dt.year
df['month']       = df['order_date'].dt.month
df['day_of_week'] = df['order_date'].dt.day_name() # e.g. "Monday", "Tuesday"
```
