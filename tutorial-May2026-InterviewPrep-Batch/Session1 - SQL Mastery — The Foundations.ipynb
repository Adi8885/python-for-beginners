## 🧠 Part 1: SQL Table Thinking (10 mins)
Before writing code, you must visualize the data.

### 🧱 The Anatomy of a Table
| Concept | Real-World Analogy |
| :--- | :--- |
| **Table** | A specific Excel Sheet (e.g., "Customers") |
| **Row** | A single record (e.g., "One specific customer") |
| **Column** | A specific attribute (e.g., "Customer Name") |

### 🛠️ Example Schema
```sql
-- Table 1: customers
-- Columns: customer_id, name, country, signup_date

-- Table 2: orders
-- Columns: order_id, customer_id, order_date, amount, status
```

---

## 🔍 Part 2: SELECT, WHERE & Sorting (25 mins)
How to "ask" the database for specific information.

### 🛠️ Your Toolbox
*   `SELECT` ➡️ Which columns do you want?
*   `WHERE` ➡️ What are your filters?
*   `ORDER BY` ➡️ How should the results be sorted?
*   `LIMIT` ➡️ How many rows do you want to see?

### 🧩 Advanced Filtering Operators
| Operator | Usage | Example |
| :--- | :--- | :--- |
| `IN` | Match multiple values | `WHERE country IN ('India', 'USA')` |
| `BETWEEN` | Match a range | `WHERE amount BETWEEN 100 AND 500` |
| `LIKE` | Pattern matching | `WHERE name LIKE 'A%'` (Starts with A) |
| `IS NULL` | Find empty values | `WHERE email IS NULL` |

---

## 📝 Practice Set 1: Basic Filtering
> **Target Table:** `customers`

```sql
-- 1. Find all customers from India.
SELECT * FROM customers WHERE country = 'India';

-- 2. Find customers who signed up in 2023.
SELECT * FROM customers WHERE signup_date LIKE '2023%';

-- 3. Find distinct countries from which customers signed up.
SELECT DISTINCT country FROM customers;

-- 4. Find customers whose name starts with 'A'.
SELECT * FROM customers WHERE name LIKE 'A%';

-- 5. Find customers whose country is either India, USA, or Canada.
SELECT * FROM customers WHERE country IN ('India', 'USA', 'Canada');

-- 6. Find the 10 most recent customers.
SELECT * FROM customers ORDER BY signup_date DESC LIMIT 10;
```

## 🔢 Part 3: Aggregation Basics (25 mins)
Summary statistics: turning data into insights! 📈

### 🧪 The Big 5 Functions
*   `COUNT()` ➡️ How many?
*   `SUM()` ➡️ Total?
*   `AVG()` ➡️ Average?
*   `MIN()` ➡️ Smallest?
*   `MAX()` ➡️ Largest?

### ⚠️ Interview Pro-Tip: The COUNT Difference
| Function | What it does |
| :--- | :--- |
| `COUNT(*)` | Counts **every single row** |
| `COUNT(column)` | Ignores **NULL** values |
| `COUNT(DISTINCT col)` | Counts **unique, non-null** values |

---

## 📝 Practice Set 2: Aggregations
> **Target Table:** `orders`

```sql
-- 1. Count total number of orders.
SELECT COUNT(*) FROM orders;

-- 2. Count total unique customers who placed orders.
SELECT COUNT(DISTINCT customer_id) FROM orders;

-- 3. Find total revenue.
SELECT SUM(amount) FROM orders;

-- 4. Find average order value.
SELECT AVG(amount) FROM orders;

-- 5. Find highest and lowest order amount.
SELECT MAX(amount), MIN(amount) FROM orders;

-- 6. Count successful orders only.
SELECT COUNT(*) FROM orders WHERE status = 'Success';
```

---

## 👯 Part 4: GROUP BY and HAVING (30 mins)
The "Secret Sauce" of SQL. Grouping allows you to see patterns across categories.

### ⚖️ The Golden Rule
> [!TIP]
> **WHERE** filters data *before* it is grouped.  
> **HAVING** filters data *after* it is grouped.

| Clause | Purpose |
| :--- | :--- |
| `GROUP BY` | Collapses rows into buckets (e.g., Revenue per Country) |
| `HAVING` | Used to filter aggregated results (e.g., Groups with > 5 orders) |

---

## 📝 Practice Set 3: Grouping
> **Target Table:** `orders`

```sql
-- 1. Find total revenue by customer.
SELECT customer_id, SUM(amount) FROM orders GROUP BY customer_id;

-- 2. Count orders by status.
SELECT status, COUNT(*) FROM orders GROUP BY status;

-- 3. Find average order value by customer.
SELECT customer_id, AVG(amount) FROM orders GROUP BY customer_id;

-- 4. Find customers with total spending greater than ₹10,000.
SELECT customer_id, SUM(amount) FROM orders GROUP BY customer_id HAVING SUM(amount) > 10000;

-- 5. Find customers who placed more than 5 orders.
SELECT customer_id, COUNT(*) FROM orders GROUP BY customer_id HAVING COUNT(*) > 5;
```
