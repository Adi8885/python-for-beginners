## 🧠 Part 1: SQL Table Thinking (10 mins)
Before writing code, you must visualize the data.

### 🧱 The Anatomy of a Table
| Concept | Real-World Analogy |
| :--- | :--- |
| **Table** | A specific Movie Catalog (e.g., "**movies**") |
| **Row** | A single record (e.g., "**A Matter of Life and Death**") |
| **Column** | A specific attribute (e.g., "**worldwide_gross_income**") |

### 🛠️ Example Schema
```sql
-- Table 1: movies
-- Columns: id, title, year, date_published, duration, country, worldwide_gross_income, languages, production_company

-- Table 2: ratings
-- Columns: movie_id, avg_rating, total_votes, median_rating
```

---


## 🔍 Part 2: SELECT, WHERE & Sorting (25 mins)
How to "ask" the database for specific information.

### 🛠️ Your Toolbox
*   `SELECT` ➡️ Which columns do you want?
*   `WHERE` ➡️ What are your filters?
*   `ORDER BY` ➡️ How should the results be sorted?
*   `LIMIT` ➡️ How many rows do you want to see?

### 🧩 Advanced Filtering Operators
| Operator | Usage | Example |
| :--- | :--- | :--- |
| `IN` | Match multiple values | `WHERE languages IN ('English', 'Hindi')` |
| `BETWEEN` | Match a range | `WHERE duration BETWEEN 80 AND 100` |
| `LIKE` | Pattern matching | `WHERE name LIKE 'A%'` (Starts with A) |
| `IS NULL` | Find empty values | `WHERE worlwide_gross_income IS NULL` |

---

**
## 📝 Practice Set 1: Basic Filtering**
> **Target Table:** `movies`
```sql
-- 1. Find all movies produced in India.
SELECT * FROM movies WHERE country = 'India';

-- 2. Find movies released in the year 2017.
SELECT * FROM movies WHERE year = 2017;

-- 3. Find movies with a duration longer than 150 minutes.
SELECT title, duration FROM movies WHERE duration > 150;

-- 4. Find movies whose title starts with 'The'.
SELECT * FROM movies WHERE title LIKE 'The%';
```

## 🔢 Part 3: Aggregation Basics (25 mins)
Summary statistics: turning data into insights! 📈

### 🧪 The Big 5 Functions
*   `COUNT()` ➡️ How many?
*   `SUM()` ➡️ Total?
*   `AVG()` ➡️ Average?
*   `MIN()` ➡️ Smallest?
*   `MAX()` ➡️ Largest?

### ⚠️ Interview Pro-Tip: The COUNT Difference
| Function | What it does |
| :--- | :--- |
| `COUNT(*)` | Counts **every single row** |
| `COUNT(column)` | Ignores **NULL** values |
| `COUNT(DISTINCT col)` | Counts **unique, non-null** values |

---

## 📝 Practice Set 2: Aggregations
> **Target Table:** `movies`

```sql
-- 1. Find the total number of movies in the database.
SELECT COUNT(*) FROM movies;

-- 2. Find the average duration of all movies.
SELECT AVG(duration) FROM movies;

-- 3. Find the highest worldwide gross income recorded.
SELECT MAX(worldwide_gross_income) FROM movies;
```

---

## 👯 Part 4: GROUP BY and HAVING (30 mins)
The "Secret Sauce" of SQL. Grouping allows you to see patterns across categories.

### ⚖️ The Golden Rule
> [!TIP]
> **WHERE** filters data *before* it is grouped.  
> **HAVING** filters data *after* it is grouped.

| Clause | Purpose |
| :--- | :--- |
| `GROUP BY` | Collapses rows into buckets (e.g., Revenue per Country) |
| `HAVING` | Used to filter aggregated results (e.g., movies with > $10,000 income) |

---

## 📝 Practice Set 3: Grouping
> **Target Table:** `movies`

```sql
-- 1. Count how many movies were released each year.
SELECT year, COUNT(*) FROM movies GROUP BY year;

-- 2. Find the number of movies produced by each production company.
SELECT production_company, COUNT(*) 
FROM movies 
GROUP BY production_company 
HAVING COUNT(*) > 10; -- Only show major companies
```
