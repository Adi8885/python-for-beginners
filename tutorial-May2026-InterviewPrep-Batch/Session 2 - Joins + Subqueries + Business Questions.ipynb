## 🌟 Why This Session?
In real-world data science, answers are rarely found in a single table. Most SQL interview questions require data from multiple sources. **Joins** are the "glue" that brings related data together, and they are the most frequently tested topic in SQL interviews! 💼

## 🧩 Part 1: Why Joins Are Needed — 10 mins

Let’s use our **IMDb Dataset** to understand why we can't just put everything in one table.

### 📁 The Data Structure
If we want to know which **Director** directed which **Movie**, we have to look at three different tables:

#### **1. `Movie` Table** 🎥
*Stores basic info about the film.*

| movie_id (PK) | title | year | rating |
| :--- | :--- | :--- | :--- |
| `tt2388771` | Mowgli | 2018 | 6.6 |
| `tt5164214` | Ocean's Eight | 2018 | 6.2 |

#### **2. `Person` Table** 👤
*Stores details about people (Actors, Directors, etc.).*

| person_id (PK) | name | gender |
| :--- | :--- | :--- |
| `nm0785227` | Andy Serkis | Male |
| `nm0002657` | Gary Ross | Male |

#### **3. `M_Director` Table** 🔗
*The "Bridge" table that connects Movies to Directors.*

| movie_id (FK) | person_id (FK) |
| :--- | :--- |
| `tt2388771` | `nm0785227` |
| `tt5164214` | `nm0002657` |

---

### 💡 The Problem
If a student wants to find: **"What is the name of the director for the movie 'Mowgli'?"**
1. The **Movie Title** is in the `Movie` table.
2. The **Director Name** is in the `Person` table.
3. Neither table has both!

### 🔑 The Solution: The "Join"
We use the **IDs** to connect these tables. `movie_id` connects `Movie` to `M_Director`, and `person_id` connects `M_Director` to `Person`.

```sql
-- Conceptual Join to get Director Names for Movies
SELECT 
    m.title, 
    p.name AS director_name
FROM Movie m
JOIN M_Director d ON m.movie_id = d.movie_id
JOIN Person p ON d.person_id = p.person_id;
```

**Why do it this way?**
- 🚀 **Efficiency:** We don't have to repeat the director's name 100 times if they directed 100 movies.
- 🧹 **Cleanliness:** Data stays organized and easy to update in one place.

## 🤝 Part 2: INNER JOIN and LEFT JOIN — 30 mins

In interviews, 90% of your Join questions will involve these two. Let's master them using the IMDb data!

### 1. The `INNER JOIN` (The Matchmaker ❤️)
The **INNER JOIN** returns only the rows where there is a match in **both** tables. If a movie doesn't have a director listed, or a director hasn't directed any movies, they won't appear here.

**Goal:** List all movies along with their Genre names.

```sql
SELECT 
    m.title, 
    g.Name AS genre_name
FROM Movie m
INNER JOIN M_Genre mg ON m.movie_id = mg.movie_id
INNER JOIN Genre g ON mg.genre_id = g.genre_id
LIMIT 5;
```

| title | genre_name |
| :--- | :--- |
| Mowgli | Adventure |
| Mowgli | Drama |
| Ocean's Eight | Action |
| Ocean's Eight | Comedy |

---

### 2. The `LEFT JOIN` (The Inclusive One 🫂)
The **LEFT JOIN** returns **all** rows from the left table, and the matched rows from the right table. If there is no match, the result is `NULL` on the right side.

**Goal:** List all Movies and their Production companies (even if no production company is listed).

```sql
-- Assuming a scenario where some movies might not have a producer entry
SELECT 
    m.title, 
    p.Name AS producer_name
FROM Movie m
LEFT JOIN M_Producer mp ON m.movie_id = mp.movie_id
LEFT JOIN Person p ON mp.person_id = p.person_id
LIMIT 5;
```

> [!TIP]
> **When to use LEFT JOIN?** Use it when you want to find "Missing" data. For example: "Find movies that have **no** director assigned." (Where `director_id IS NULL`).

---

### 📝 Summary Table

| Join Type | What it returns | Emoji |
| :--- | :--- | :--- |
| **INNER JOIN** | Only matching records from both tables. | 🤝 |
| **LEFT JOIN** | All records from Left + Matches from Right. | 👈 |
| **RIGHT JOIN** | All records from Right + Matches from Left. (Rarely used) | 👉 |
| **FULL JOIN** | All records from both tables. (Rarely used) | 🌍 |

## 📈 Part 3: Join + Group By Interview Problems — 25 mins

This is where SQL gets powerful! In technical interviews, you aren't just asked to "join tables"—you're asked to **analyze data** across them.

### ⚠️ The "Duplicate Trap" (Interview Alert!)
When you join a table that has multiple matches (e.g., a Movie has 3 Genres), the rows multiply. If you join `Movie` to `M_Genre` AND `M_Cast`, one movie might appear 15 times!
*   **The Fix:** Always use `COUNT(DISTINCT column_name)` instead of `COUNT(*)` when counting items after a complex join.

---

### 🚀 Top 6 Interview Problems (IMDb Edition)

#### **1. The Genre Specialist** 🎭
*Question:* Which genre has the highest average movie rating?
*Tables:* `Movie`, `M_Genre`, `Genre`

```sql
SELECT 
    g.Name AS genre_name, 
    ROUND(AVG(m.rating), 2) AS avg_rating,
    COUNT(m.movie_id) AS movie_count
FROM Movie AS m
JOIN M_Genre AS mg ON m.movie_id = mg.movie_id
JOIN Genre AS g ON mg.genre_id = g.genre_id
GROUP BY g.Name
ORDER BY avg_rating DESC;
```

#### **2. The Most Prolific Director** 🎬
*Question:* List the top 5 directors who have directed the most movies, along with their average rating.
*Tables:* `Movie`, `M_Director`, `Person`

```sql
SELECT 
    p.Name AS director_name, 
    COUNT(m.movie_id) AS total_movies,
    AVG(m.rating) AS avg_director_rating
FROM Person AS p
JOIN M_Director AS d ON p.person_id = d.person_id
JOIN Movie AS m ON d.movie_id = m.movie_id
GROUP BY p.Name
HAVING total_movies > 1
ORDER BY total_movies DESC
LIMIT 5;
```

#### **3. Multi-Language Mastery** 🌎
*Question:* Which language has the most 'Action' movies?
*Tables:* `Movie`, `M_Language`, `Language`, `M_Genre`, `Genre` (4 Tables!)

```sql
SELECT 
    l.Name AS language_name, 
    COUNT(DISTINCT m.movie_id) AS action_movie_count
FROM Movie AS m
JOIN M_Language AS ml ON m.movie_id = ml.movie_id
JOIN Language AS l ON ml.language_id = l.language_id
JOIN M_Genre AS mg ON m.movie_id = mg.movie_id
JOIN Genre AS g ON mg.genre_id = g.genre_id
WHERE g.Name = 'Action'
GROUP BY l.Name
ORDER BY action_movie_count DESC;
```

#### **4. Global Box Office Influence** 🗺️
*Question:* Which country has produced movies with the highest total number of votes?
*Tables:* `Movie`, `M_Country`, `Country`

```sql
SELECT 
    c.Name AS country_name, 
    SUM(m.votes) AS total_votes
FROM Movie AS m
JOIN M_Country AS mc ON m.movie_id = mc.movie_id
JOIN Country AS c ON mc.country_id = c.country_id
GROUP BY c.Name
ORDER BY total_votes DESC;
```

#### **5. Cast Heavyweights** 👯
*Question:* Which movie has the largest cast (most actors listed)?
*Tables:* `Movie`, `M_Cast`

```sql
SELECT 
    m.title, 
    COUNT(mc.person_id) AS cast_count
FROM Movie AS m
JOIN M_Cast AS mc ON m.movie_id = mc.movie_id
GROUP BY m.title
ORDER BY cast_count DESC
LIMIT 1;
```

#### **6. The "Success" Duo** 🤝
*Question:* Find the names of actors who have worked in movies with a rating higher than 8.0.
*Tables:* `Movie`, `M_Cast`, `Person`

```sql
SELECT 
    DISTINCT p.Name AS actor_name
FROM Person AS p
JOIN M_Cast AS mc ON p.person_id = mc.person_id
JOIN Movie AS m ON mc.movie_id = m.movie_id
WHERE m.rating > 8.0
ORDER BY p.Name ASC;
```

---

### 💡 Practice Checklist for Students:
1.  **Use Aliases:** Always use `AS m`, `AS p`, etc. to keep code readable.
2.  **Filter First:** Use `WHERE` to reduce rows *before* grouping if possible.
3.  **Validate:** Does the result make sense? (If you see 1,000 actors for one movie, you probably missed a join condition!)

## ⚠️ Part 4: Common Join Traps — 10 mins

Interviewer: *"Your query looks correct, but the numbers are way off. Do you know why?"* 
This is the moment they test if you understand how Joins **actually** work behind the scenes.

---

### 🪤 Trap 1: The "Where" Clause that Kills the "Left Join"

**1. The Trap:**
Trying to find all movies and their actors, but filtering for only Female actors in the `WHERE` clause.
```sql
SELECT m.title, p.Name
FROM Movie m
LEFT JOIN M_Cast mc ON m.movie_id = mc.movie_id
LEFT JOIN Person p ON mc.person_id = p.person_id
WHERE p.gender = 'Female'; -- ❌ This turns it back into an INNER JOIN!
```

**2. Explain the Trap:**
A `LEFT JOIN` keeps all movies even if they have no actors. But the `WHERE` clause runs **after** the join. Since a movie with no actors will have `NULL` for gender, and `NULL = 'Female'` is false, those movies are deleted from the result.

**3. Better Solution:**
Move the condition into the `ON` clause or handle NULLs.
```sql
SELECT m.title, p.Name
FROM Movie m
LEFT JOIN M_Cast mc ON m.movie_id = mc.movie_id
LEFT JOIN Person p ON mc.person_id = p.person_id AND p.gender = 'Female'; -- ✅ Correct
```

---

### 🪤 Trap 2: The Fan-Out (Duplicate Counting)

**1. The Trap:**
Trying to count how many Genres and how many Actors a movie has in a single query.
```sql
SELECT m.title, COUNT(mg.genre_id) as g_count, COUNT(mc.person_id) as a_count
FROM Movie m
JOIN M_Genre mg ON m.movie_id = mg.movie_id
JOIN M_Cast mc ON m.movie_id = mc.movie_id
GROUP BY m.title; -- ❌ Both counts will be the same (and wrong!)
```

**2. Explain the Trap:**
If a movie has 2 genres and 10 actors, SQL creates **20 rows** (2x10) for that movie. `COUNT` will return 20 for both columns.

**3. Better Solution:**
Always use `DISTINCT` when counting across multiple joined tables.
```sql
SELECT m.title, COUNT(DISTINCT mg.genre_id), COUNT(DISTINCT mc.person_id)
FROM Movie m
...
```

---

### 🪤 Trap 3: Joining on Non-Unique Columns

**1. The Trap:**
Joining tables using "Name" instead of "ID".
```sql
SELECT m.title, p.Name
FROM Movie m
JOIN Person p ON m.director_name = p.Name; -- ❌ Trap!
```

**2. Explain the Trap:**
Names are not unique. If there are two directors named "James Cameron", SQL will join the movie to **both** of them, doubling your rows and your budget!

**3. Better Solution:**
Always join on **Primary Keys (PK)** and **Foreign Keys (FK)**.
```sql
JOIN M_Director md ON m.movie_id = md.movie_id
JOIN Person p ON md.person_id = p.person_id; -- ✅ Safe
```

---

### 🪤 Trap 4: The Accidental Cross Join (Cartesian Product)

**1. The Trap:**
Forgetting the `ON` clause or using commas in the `FROM` statement.
```sql
SELECT m.title, g.Name
FROM Movie m, Genre g; -- ❌ Disaster!
```

**2. Explain the Trap:**
Without a join condition, SQL matches **every single movie** with **every single genre**. If you have 1,000 movies and 20 genres, you get 20,000 rows of garbage data.

**3. Better Solution:**
Use explicit `JOIN ... ON` syntax always.
```sql
FROM Movie m JOIN M_Genre mg ON m.movie_id = mg.movie_id ...
```

---

### 🪤 Trap 5: The "Inner Join" Data Loss

**1. The Trap:**
Using `INNER JOIN` to fetch all movies and their ratings.
```sql
SELECT m.title, m.rating
FROM Movie m
INNER JOIN M_Director md ON m.movie_id = md.movie_id; -- ❌ Data Loss!
```

**2. Explain the Trap:**
If a movie is brand new and doesn't have a director assigned yet in the `M_Director` table, an `INNER JOIN` will **delete** that movie from your list entirely. 

**3. Better Solution:**
Use `LEFT JOIN` if the requirement is to show **all movies**, regardless of whether they have a director.

---

### 💡 Summary for Students:
*   **Counting?** Use `DISTINCT`.
*   **Filtering Right Table?** Be careful with `WHERE`.
*   **Joining?** Use IDs, not Names.
*   **All items needed?** Use `LEFT JOIN`.

## 🏗️ Part 5: Subqueries and CTEs — 30 mins

Sometimes, a single `JOIN` isn't enough. You might need to find an answer based on *another* answer. This is where **Subqueries** and **CTEs** (Common Table Expressions) come in.

### 🍱 What are they?
*   **Subquery:** A query nested inside another query. Think of it as a "Query inception." 🌀
*   **CTE (`WITH`):** A way to create a temporary, named result set that you can use like a regular table. It makes your code look much cleaner! ✨

---

### 🚀 6 Practical Examples (IMDb Edition)

#### **Example 1: The "Above Average" Filter (Subquery in `WHERE`)**
*Goal:* Find movies that have a rating higher than the average rating of all movies.

```sql
SELECT title, rating
FROM Movie
WHERE rating > (SELECT AVG(rating) FROM Movie); -- The subquery runs first!
```

#### **Example 2: Finding Specialists (Subquery with `IN`)**
*Goal:* Find the names of actors who have worked with the director 'Andy Serkis'.

```sql
SELECT Name 
FROM Person 
WHERE person_id IN (
    -- Subquery finds IDs of movies directed by Andy Serkis
    SELECT movie_id FROM M_Director 
    WHERE person_id = (SELECT person_id FROM Person WHERE Name LIKE '%Andy Serkis%')
);
```

#### **Example 3: The Temporary Result (Subquery in `FROM`)**
*Goal:* Treat the average rating per year as a table and find years where the average was high.

```sql
SELECT year, avg_yr_rating
FROM (
    SELECT year, AVG(rating) AS avg_yr_rating
    FROM Movie
    GROUP BY year
) AS yearly_stats
WHERE avg_yr_rating > 7.5;
```

#### **Example 4: The Clean Way (CTE using `WITH`)**
*Goal:* Same as Example 3, but much easier to read!

```sql
WITH YearlyStats AS (
    SELECT year, AVG(rating) AS avg_yr_rating
    FROM Movie
    GROUP BY year
)
SELECT * 
FROM YearlyStats 
WHERE avg_yr_rating > 7.5;
```

#### **Example 5: Multi-Step Analysis (Multiple CTEs)**
*Goal:* Find directors who have directed movies in the 'Action' genre that are rated above 8.0.

```sql
WITH ActionMovies AS (
    SELECT m.movie_id, m.title
    FROM Movie m
    JOIN M_Genre mg ON m.movie_id = mg.movie_id
    JOIN Genre g ON mg.genre_id = g.genre_id
    WHERE g.Name = 'Action' AND m.rating > 8.0
),
ActionDirectors AS (
    SELECT md.person_id, am.title
    FROM M_Director md
    JOIN ActionMovies am ON md.movie_id = am.movie_id
)
SELECT p.Name, ad.title
FROM Person p
JOIN ActionDirectors ad ON p.person_id = ad.person_id;
```

#### **Example 6: Finding "Empty" Genres (Subquery in `WHERE NOT EXISTS`)**
*Goal:* Find any genres that currently have zero movies associated with them.

```sql
SELECT Name
FROM Genre g
WHERE NOT EXISTS (
    SELECT 1 FROM M_Genre mg WHERE mg.genre_id = g.genre_id
);
```

---

### ⚖️ Subquery vs. CTE: Which one should you use?

| Feature | Subquery | CTE (`WITH`) |
| :--- | :--- | :--- |
| **Readability** | 📉 Hard to read if nested deep. | 📈 Very clean and sequential. |
| **Reusability** | ❌ Can only be used once. | ✅ Can be referenced multiple times in the same query. |
| **Organization** | 📦 Hidden inside clauses. | 🏗️ Defined at the top like a variable. |
| **Interview Tip** | Fine for simple filters. | **Always use CTEs** for complex multi-step problems to impress interviewers! |

---
**💡 Pro Tip:** If your query has more than 2 levels of nesting, **convert it to a CTE**. Your future self (and your interviewer) will thank you! 👏

### 🛠️ Need a Starting Point?

1.  `FROM` & `JOIN` (Gather your data)
2.  `WHERE` (Filter the rows)

4.  `HAVING` (Filter the groups)
5.  `SELECT` (Pick your columns)
6.  `ORDER BY` (Sort the results)
