## Loops

- The "Why": Computers are great at doing boring, repetitive tasks without getting tired. Loops allow us to run the same block of code multiple times.

- For Loops: Use these when you know how many times you want to repeat something.

- While Loops: Use these when you want to repeat something until a condition changes (like waiting for a user to type "quit").

- The Danger Zone: Mention "Infinite Loops"—it’s a rite of passage for every programmer!

### For Loop

```python
# The For Loop: Counting to 5
# range(1, 6) starts at 1 and stops BEFORE 6
print("Counting:")
for i in range(1, 6):
    print(i)
```

### While Loop

```python
# The While Loop: Controlled by a condition
count = 5
while count > 0:
    print(f"Countdown: {count}")
    count = count - 1
print("Take off!")
```
